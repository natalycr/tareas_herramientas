```python
%pylab inline

from scipy.optimize import curve_fit
```

```python
def R2(xExp, xTeo):
    EE= sum((xTeo- xExp)**2)
    mean = sum(xExp)/(len(xExp))
    variance = sum((mean-xExp)**2)
    R2 = 1 - EE/variance
    return R2
```

```python
data1 = loadtxt("data1.dat")
x = data1[:,0] # sacamos los  datos de data en la  columna 0
y = data1[:,1]

scatter( x, y, c='g', s=40, label="Datos 1")

# para la linea 

def expfunc(x,a,b):
    return a*e**(x-b)
fitpars, covmat = curve_fit(expfunc,x,y)

plot(x, expfunc(x, fitpars[0], fitpars[1]), 'r-', label="Regrecion")

# Para cada grafica

title("$Datos\ 1$", fontsize=20)
xlabel("$X$", fontsize=20)
ylabel("$Y$", fontsize=20, rotation="horizontal")
legend(loc="best")
text(-1, 8000, "$Eq.=$"+str(fitpars[0])+"$e^(x-$"+str(fitpars[1])+"$)$", fontsize=10)
```

```python
yTeo= fitpars[0]*e**(x-fitpars[1])
R2(y, yTeo)
```

```python
yTeo= fitpars[0]*x**(fitpars[1])
R2(y, yTeo)
```
