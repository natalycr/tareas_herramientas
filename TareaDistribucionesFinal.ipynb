```python
def listafrac(M,N):
    
    lista = []
    caras = []
    sellos = []
        
    for j in range (M):#recorre el numero de muestras 
        car=0.0
        
        for i in range (N): #recorre  el numero  de  lanzamientos por muestras 
            x=np.random.randint(2)
            if  (x==1):
                car=car+1 # suma si  es cara
        lista.append(car/N)
        caras.append(car)
        sell=N-car
        sellos.append(sell)
        
    return lista, caras, sellos
```

```python
# Tarea  2 / graf  1
#donde M numero de muestras, i= numero de lanzamientos que varia , M2=
#i = 1000
#M2 = 20000
def resta(M,i, M2):
    datosplo = []
    ejex=[]
    while i <= M2: 
        x ,y , z= listafrac(M, i)
        y_m = mean(y)#media de M datos de nuero de caras por i lanzamientos
        z_m = mean(z)#media de M datos de nuero de sellos por i lanzamientos 
        dato=abs(y_m-z_m)
        datosplo.append(dato)
        ejex.append(i)
        i=i+1000
    return ejex, datosplo
```

```python
x1, y1=resta(20, 1000, 20000)
x2, y2=resta(40, 1000, 20000)
scatter(x1 , y1, s=35, c='r')
scatter(x2 , y2, s=35, c='G', marker='^')

title("$Grafica\ diferencias\ de\ medias\ para\ 20\ y\ 40\ muestras$", fontsize=25)
xlabel("$Numero\ de\ lanzamientos\ por\ muestra$", fontsize=20)
ylabel("Diferencia de medias \n numero de caras-sellos", fontsize=15, rotation="horizontal")

# axes.legend(loc=2)#legenda en la  esquina  superior izquierda

#Donde  los trianfgulos  representan los  datos  con  40  muestras  y los  circulos de 20  muestras
```

```python
#fracion caras/sellos
def fraccion(M,i, M2):
    datosplo = []
    ejex=[]
    while i <= M2: 
        x ,y , z= listafrac(M, i)
        y_m = mean(y)
        z_m = mean(z) 
        dato=(y_m/z_m)
        datosplo.append(dato)
        ejex.append(i)
        i=i+1000
    return ejex, datosplo
```

```python
x1, y1=fraccion(20, 1000, 20000)
x2, y2=fraccion(40, 1000, 20000)
scatter(x1 , y1, s=40, c='r')
scatter(x2 , y2, s=40, c='G', marker='^')


title("$Grafica\ fraccion\ de\ medias\ para\ 20\ y\ 40\ muestras$", fontsize=25)
xlabel("$Numero\ de\ lanzamientos\ por\ muestra$", fontsize=25)
ylabel("Fraccion de medias \n numero de caras/sellos", fontsize=15, rotation="horizontal")
#axes.legend(loc=4)# Legenda esquina inferior derecha
```
